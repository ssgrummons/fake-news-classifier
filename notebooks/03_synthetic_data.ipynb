```python
import random
import pandas as pd

# Define example post generators
bullshit_topics = [
    "vibe coding", "insurance disruption", "AI-generated investing tips", 
    "crypto emotional intelligence", "zero-trust magic security", "remote work as a divine calling"
]

substantive_topics = [
    "challenges of deploying LLMs in prod", "how to tune hyperparameters", 
    "designing a zero-trust architecture", "measuring claim risk in insurance", 
    "real-world usage of LangChain", "challenges of building secure crypto wallets"
]

def generate_bullshit_post(topic):
    openers = [
        f"I just discovered something that blew my mind about {topic}.",
        f"I can't believe more people aren’t talking about this {topic} hack.",
        f"This will disrupt everything you thought you knew about {topic}.",
        f"I built a full-stack app for {topic} in under 10 minutes with just 3 prompts."
    ]
    closers = [
        "Share this with someone who still writes code manually.",
        "If you’re not using this, you’re already behind.",
        "This is the future. Don’t get left behind.",
        "I’ll be writing more about this—follow for updates!"
    ]
    content = (
        f"{random.choice(openers)}\n\n"
        f"Here’s how:\n"
        f"✅ Step 1: Prompt the AI to do the heavy lifting\n"
        f"✅ Step 2: Refine the output a bit\n"
        f"✅ Step 3: Launch it live\n\n"
        f"{random.choice(closers)}"
    )
    return content.strip()

def generate_substantive_post(topic):
    intros = [
        f"Here are three hard lessons I learned about {topic}.",
        f"Let’s talk about what no one tells you about {topic}.",
        f"Here’s a quick breakdown of what actually works in {topic}.",
        f"{topic.capitalize()} is not as simple as it sounds. Here’s what I found."
    ]
    conclusions = [
        "The devil is in the details, and tradeoffs matter.",
        "This took weeks to get right—and it still breaks sometimes.",
        "Real-world complexity always kicks in past the first prototype.",
        "What worked in a blog post failed in production. Context matters."
    ]
    content = (
        f"{random.choice(intros)}\n\n"
        f"- You can’t skip error handling or fallbacks\n"
        f"- Tooling and documentation are rarely aligned\n"
        f"- Benchmarks lie; test on your real use case\n\n"
        f"{random.choice(conclusions)}"
    )
    return content.strip()

# Generate dataset
random.seed(42)
bullshit_posts = [generate_bullshit_post(random.choice(bullshit_topics)) for _ in range(25)]
substantive_posts = [generate_substantive_post(random.choice(substantive_topics)) for _ in range(25)]

df = pd.DataFrame({
    "text": bullshit_posts + substantive_posts,
    "is_bs": [1]*25 + [0]*25
})

df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle
df.head()
```

```python
df.to_csv("../data/raw/synthetic_data.csv", index=False)
```
